## ***IMPORT LIBRARIES***

```python
import cv2
import mediapipe as mpa
import numpy as np
import matplotlib.pyplot as plt
```

# ***TASKS:***

## ***1.  Pose Detection & Pre-processing***

```python
# MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()

mp_draw = mp.solutions.drawing_utils
```

```python
# LOAD VIDEO
video_path = "/content/6002033-hd_720_1280_30fps.mp4"
cap = cv2.VideoCapture(video_path)
```

```python
# GET VIDEO PROPERTIES

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
```

```python
# OUTPUT VIDEO
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

out = cv2.VideoWriter(
    '/content/output.mp4',
    fourcc,
    fps,
    (width, height)
)
```

```python
# PREVIOUS HIP POSITION
prev_hip_y = None
```

## ***2.  Joint Angle Computation & Tracking***

```python
# ANGLE FUNCTION

def calculate_angle(a, b, c):

    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(
        c[1] - b[1],
        c[0] - b[0]
    ) - np.arctan2(
        a[1] - b[1],
        a[0] - b[0]
    )

    angle = np.abs(radians * 180.0 / np.pi)

    if angle > 180:
        angle = 360 - angle

    return angle
```

## ***3.  Rule-Based Activity Classification***

```python
# PROCESS VIDEO

while cap.isOpened():

    ret, frame = cap.read()

    if not ret:
        break

    # Convert BGR to RGB
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Pose Detection
    results = pose.process(rgb)

    activity = "No Person Detected"

    if results.pose_landmarks:

        landmarks = results.pose_landmarks.landmark

        # LEFT LEG
        left_hip = [
            landmarks[23].x,
            landmarks[23].y
        ]

        left_knee = [
            landmarks[25].x,
            landmarks[25].y
        ]

        left_ankle = [
            landmarks[27].x,
            landmarks[27].y
        ]

        # RIGHT LEG
        right_hip = [
            landmarks[24].x,
            landmarks[24].y
        ]

        right_knee = [
            landmarks[26].x,
            landmarks[26].y
        ]

        right_ankle = [
            landmarks[28].x,
            landmarks[28].y
        ]

        # CALCULATE KNEE ANGLES
        left_angle = calculate_angle(
            left_hip,
            left_knee,
            left_ankle
        )

        right_angle = calculate_angle(
            right_hip,
            right_knee,
            right_ankle
        )

        avg_angle = (left_angle + right_angle) / 2

        # HIP CENTER POSITION
        hip_y = (
            landmarks[23].y +
            landmarks[24].y
        ) / 2

        # MOVEMENT DETECTION
        if prev_hip_y is not None:

            movement = hip_y - prev_hip_y

            if movement > 0.003:
                activity = "Descending"

            elif avg_angle > 150:
                activity = "Walking"

            else:
                activity = "Transition"

        prev_hip_y = hip_y

        # DRAW LANDMARKS
        mp_draw.draw_landmarks(
            frame,
            results.pose_landmarks,
            mp_pose.POSE_CONNECTIONS
        )

        # DISPLAY ACTIVITY
        cv2.putText(
            frame,
            f'Activity: {activity}',
            (30, 50),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 255, 0),
            2
        )

        # DISPLAY ANGLES
        cv2.putText(
            frame,
            f'Left Knee: {int(left_angle)}',
            (30, 100),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.8,
            (255, 255, 0),
            2
        )

        cv2.putText(
            frame,
            f'Right Knee: {int(right_angle)}',
            (30, 140),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.8,
            (255, 255, 0),
            2
        )

    # WRITE FRAME
    out.write(frame)
```

SAVING VIDEO

```python
# RELEASE EVERYTHING
cap.release()
out.release()

print("Processing Completed")
print("Output Video Saved Successfully")
print("/content/output.mp4")
```
